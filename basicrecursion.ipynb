```python
############################################################
# Recursion.py 
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  test factorial
###########################################################    
class Recursion():
    def __init__(self):
        #Nothing can be added here
        pass

    ############################################################
    #          WRITE ALL YOUR PUBLIC FUNCTION BELOW
    ########################################################### 
    def FibR(self,n:'int'):
        if n<2:
            return n
        return self.FibI(n-1)+ self.FibI(n-2)

    def FibI(self,n:'int'):
        if n<2:
            return n
        secondLast=0
        Last=1
        ans=0
        for i in range(2,n+1):
            ans=Last+secondLast
            secondLast=Last
            Last=ans
        return ans

    def thR(self, n:'int'):
        self.thrMoves(n)
        a=self.thRCount(n)
        return a


    def thrMoves(self,n:'int',s=1,t=2,d=3):
        if n==0:
            return
        if n==1:
            print(s, '->', d)
            return
        else:
            self.thrMoves(n-1,s,d,t)
            print(s, '->', d)
            self.thrMoves(n-1,t,s,d)

    def thRCount(self,n:'int',s=1,t=2,d=3):
        count=0
        if n==0:
            return 0
        elif n==1:
            return 1
        else:
           count +=self.thRCount(n-1,s,d,t)
           count +=self.thRCount(1,s,d,t)
           count +=self.thRCount(n-1,t,s,d)
        return count



    def Combined(self,n:'int'):
        a=self._permR(self._fill(n))
        count=1
        for i in a:
            if i==[0] or i==[]:
                return
            else:
                print (count ,':' , i)
                count =count+1

    def permR(self,n:'int'):
        self.Combined(n)
        a=self.permRCount(n)
        return a
        
    def permRCount(self,n:'int'):
        if n==0:
            return 1
        else:
            return n* self.permRCount(n-1)
            

    def _fill(self,n:'int'):
        aList=[]
        for i in range (n):
            aList.append(i)
        return aList

    def _permR( self,nums:'list'):
        if len(nums) == 1:
            return [nums]
        result = []
        for i in range(len(nums)):
            others = nums[:i] + nums[i+1:]
            other_permutations = self._permR(others)
            for permutation in other_permutations:
                result.append([nums[i]] + permutation)
        return result



    ############################################################
    #          WRITE ALL YOUR PRIVATE FUNCTION BELOW
    ########################################################### 
   
```
